```python
from setfit import AbsaTrainer, TrainingArguments, AbsaModel
from datasets import load_dataset
from sklearn.metrics import f1_score, accuracy_score
```

```python
model = AbsaModel.from_pretrained(
    "/kaggle/input/setfit-fiqa/checkpoints/setfit-absa-model-aspect",
    "/kaggle/input/setfit-fiqa/checkpoints/setfit-absa-model-polarity",
    spacy_model="en_core_web_lg",
)
```

```python
from datasets import load_dataset
```

```python
dataset = load_dataset("ronenlap/SetFitAbsa_FiQA")
train_ds = dataset["train"]
test_ds = dataset["test"]
```

```python
test_ds = test_ds.rename_columns({"sentence": "text","aspect": "span"})
```

```python
output  = model.predict(test_ds) # a new column which holds the predicted polarity, "pred_polarity", is added to the dataset
output
```

```python
weighted_f1_score = f1_score(output["polarity"], output["pred_polarity"], average="weighted")
print(f"weighted_f1_score: {weighted_f1_score}")
```

```python
accuracy = accuracy_score(output["polarity"], output["pred_polarity"])
print(f"accuracy_score: {accuracy}")
```

```python
import json
import pandas as pd
with open("/kaggle/working/fiqa2.json", "r") as file:
    data = json.load(file)

# Create a list to store each row's information
rows = []

# Extract data for each entry
for item in data.values():
    sentence = item["sentence"]
    for info in item["annotation"]:
        rows.append({
            "sentence": sentence,
            "target": info["target"],
            "sentiment": info["sentiment"],
            "aspect": info["aspect"]
        })

# Convert to DataFrame
df = pd.DataFrame(rows, columns=["sentence", "target", "sentiment", "aspect"])

print(df)
```
